```python
import pandas as pd
import csv
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt 
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
from statistics import mean
import sklearn as sk
from numpy import array
```

```python
#creating my dataframe
mydata=pd.DataFrame({"Db": [], "Hz": []})
```

```python
#filling the dataframe with the file from SJapp
#i have all files in the same directory as .py file, they are named progressively with the sintax:
#   PROJECTNAME Registrazione n.csv    with n in [1,2...]

#my project name is sl, so i run a for cycle within the number of record i have and fill my dataframe
for i in range(1,27):
    cmd='pd.read_csv("sl Registrazione '+str(i)+'.csv")' 
    df=eval(cmd)
    
    Db=df["DecibelSource"].dropna()  #clean from NAs
    Hz=df["PitchSensor"].dropna()
    
    
    
    Db=list(Db)
    Hz=list(Hz)
    
    #Db=Db.reset_index(drop=True)
    #Hz=Hz.reset_index(drop=True)
    
    
    
    mydata=mydata.append({'Db' : Db , 'Hz' : Hz} ," ignore_index=True")  #write new line with the value found in the CSV
    
```

```python
#now i read the csv with the labels associated to my dataframe
labels=pd.read_csv("labels.csv", sep=";")
```

```python
#let's concatenate the two tables
result = pd.concat([mydata, labels], axis=1, sort=False)
```

# Regression 

```python
#evaluate mininum len
mini=100000000000000000
for i in range(len(result)):
    now=len(result["Db"][i])
    #print(str(i)+ " " +str(now))
    if now<mini:
        mini=now
mini
```

```python
#same len for all

for i in range(len(result)):
    result["Db"][i]=result["Db"][i][0:mini]

for i in range(len(result)):
    result["Hz"][i]=result["Hz"][i][0:mini]
```

```python
#selecting only male and female to study their properties
male=result.loc[result["Male"]==1]
male=male.reset_index(drop=True)  #setting new indexes for the new DF
female=result.loc[result["Male"]==0]
female=female.reset_index(drop=True)
```

```python
##let's see how the means differs
somma=0
for i in range(len(male)):
    somma+=mean(male["Db"][i])
print("mean for DB in male is "+str(round(somma/(len(male)+1),2)))

somma=0
for i in range(len(male)):
    somma+=mean(male["Hz"][i])
print("mean for HZ in male is " + str(round(somma/(len(male)+1),2)))

##now for female..
somma=0
for i in range(len(female)):
    somma+=mean(female["Db"][i])
print("mean for DB in female is "+str(round(somma/(len(female)+1),2)))

somma=0
for i in range(len(female)):
    somma+=mean(female["Hz"][i])
print("mean for HZ in female is " +str(round(somma/(len(female)+1),2)))
```
