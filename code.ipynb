# Title: Gender and Feeling recognition through voice tone and intensity

# Data collection & sources:

This pilot dataset shows how our data will be structured: 

Each record is stored in a row of a matrix, which colums are:

-Db: the intensity of sound during the recording of the voice, expressed in decibel, stored in a vector.

-Hz: the pitch of the voice in Hz, also stored in a vector:

-Male: boolean variable, 1 if the speaker is a male, 0 if a female

-Feeling: categorical variable that takes as value one of the 4 feelings we are dealing with

-Person: categorical variable that takes the name of the person speaking, if the speaker is part of the project developers group; otherwise has value "other"

# Software Toolkit/ Update:

While collecting the data we have decided to work on python instead of R because is easier to work with matrix having arrays as elements.

# Model & Methods/ Update:

In order to distinguish through feelings we used the model proposed by Gloria Willcox (see references):

This "Feeling wheel" helps us to associate every feeling to 6 larger group (joyful, powerful, paceful, scared, mad, sad). We used them as labels for the feeling of the speaker we were recording.

Since some feelings was never used in our dataset, or has very low frequency, we merged them together and get 4 labels: sad, happy, angry, steady( when someone is not showing any particoular feeling)

```python
import pandas as pd
import csv
```

```python
#creating my dataframe
mydata=pd.DataFrame({"Db": [], "Hz": []})
```

```python
#filling the dataframe with the file from SJapp
#i have all files in the same directory as .py file, they are named progressively with the sintax:
#   PROJECTNAME Registrazione n.csv    with n in [1,2...]

#my project name is sl, so i run a for cycle within the number of record i have and fill my dataframe
for i in range(1,27):
    cmd='pd.read_csv("sl Registrazione '+str(i)+'.csv")' 
    df=eval(cmd)
    Db=df["DecibelSource"].dropna()  #clean from NAs
    Hz=df["PitchSensor"].dropna()
    mydata=mydata.append({'Db' : Db , 'Hz' : Hz} , ignore_index=True)  #write new line with the value found in the CSV
    
```

```python
mydata
```

We want to structure our data by having a column for each variable we are going to record, that are sound intensity (Db) and pitch (Hz);
Since for each record we have a set of values, we store them into an array, so that a single row represent a single recording session, and contains all the information we need.

```python
#now i read the csv with the labels associated to my dataframe
labels=pd.read_csv("labels.csv", sep=";")
```

```python
labels
```

For each record in the science journal app we had set 3 labels to every record:

-Male, a boolean variable that has value 1 if the person that were speaking is a male, 0 if female;

-Feeling, categorical variable with the value of the feeling that the speaker has during the record;

-Person, categorical variable with the name of the project developer that was speaking (if the speaker is not part of the group the value will be "other"). We will use this value to let the model distinguish the difference of voice tone and intensity through different speakers.

```python
#let's concatenate the two tables
result = pd.concat([mydata, labels], axis=1, sort=False)
```
